Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
data = titanic.groupby(titanic['sex']).count()
plt.pie(data['name'], labels=data.index, autopct="%.2f")
plt.title('Traveller Distribution Gender Wise')
plt.show()
```

```python
titanic['gender'] = 'g'
titanic.loc[titanic['sex'] == 'male', 'gender'] = 'b'
titanic.loc[titanic['sex'] == 'female', 'gender'] = 'r'

plt.scatter(x=titanic['fare'], y=titanic['age'], c=titanic['gender'])
plt.legend(titanic['sex'])
plt.title('Fare Paid Distribution Age Wise')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
```
